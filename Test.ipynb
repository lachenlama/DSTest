```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Top_100_most_streamed.csv")
df.head()
```

```python
df.shape
```

### Pop music dataframe below

```python
df_of_pop = pd.read_csv("Top_100_most_streamed.csv", index_col='year')
df_of_pop = df_of_pop.rename(columns={'top genre': 'top_genre'})
df_of_pop.head()
```

```python
df_of_pop = df_of_pop[df_of_pop['top_genre'] == 'pop']
df_of_pop = df_of_pop.sort_index()
df_of_pop.head()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(16,6))
plt.title("Popularity of Pop Music over the years", fontsize=20)
sns.lineplot(data=df_of_pop['popularity'])
plt.ylabel("Popularity", fontsize=20)
plt.xlabel("Year", fontsize=20)
```

#### Figure above is the popularity of Pop music over the years.

---

```python
df = df.rename(columns={'top genre' : 'top_genre'})
df_of_rap = df[df['top_genre'].str.contains('rap')] # returns any str that contains the certain word
df_of_hiphop = df[df['top_genre'].str.contains('hip hop')]
```

```python
plt.figure(figsize=(16,6))
plt.title("Popularity of Rap and Hip Hop music over the years", fontsize=20)
plt.xlabel("Year", fontsize=20)
plt.ylabel("Popularity", fontsize=20)
sns.lineplot(data=df_of_rap, x='year', y='popularity', label='rap')
sns.lineplot(data=df_of_hiphop, x='year', y='popularity', label='hiphop')
```
